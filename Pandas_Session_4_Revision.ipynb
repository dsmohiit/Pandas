```python
# How to find the number of groups formed?
print(len(movies.groupby("Genre")))
print(movies["Genre"].nunique())
```

```python
# How to find number rows in each group?
print(movies.groupby("Genre").size().sort_values(ascending= False))
print(movies["Genre"].value_counts())
```

```python
# Find out the first item of each group.
movies.groupby("Genre").first()
```

```python
# Find out the last item of each group.
movies.groupby("Genre").last()
```

```python
# Find out the nth item of each group.
# 8th movie of each group.
movies.groupby("Genre").nth(8)
```

```python
# get_group() is used to retrieve a particular group.
movies.groupby("Genre").get_group("Action")
```

```python
# agg() method
movies.groupby("Genre")["IMDB_Rating"].agg(["mean", "sum", "max", "min"])
```

```python
movies.groupby("Genre").agg({
    "Runtime": "mean",
    "No_of_Votes": "sum",
    "IMDB_Rating": "mean"
})
```

```python
# split --> apply --> combine
# Find the number of movies with "A" for each movies.
def foo(group):
  return group["Series_Title"].str.startswith("A").sum()
```

```python
genre = movies.groupby("Genre")
genre.apply(foo)
```

```python
movies[movies["Series_Title"].str.startswith("A")]["Genre"].value_counts()
```

```python
# Find ranking of each movie in the group according to IMDB score.
def rank_movies(group):
  group["Rank"] = group["IMDB_Rating"].rank(ascending= False)
  return group
```

```python
movies.groupby("Genre").apply(rank_movies)
```

```python
def normalized_rating(group):
  group["Normal_Rating"] = (group["IMDB_Rating"] - group["IMDB_Rating"].min()) / (group["IMDB_Rating"].max() - group["IMDB_Rating"].min())
  return group
```

```python
movies.groupby("Genre").apply(normalized_rating)
```

```python
# groupby on multiple columns
duo = movies.groupby(["Director", "Star1"])
duo.size()
```

```python
duo.get_group(("Aamir Khan",	"Amole Gupte"))
```

```python
# Find the most earning actor director combo.
duo["Gross"].sum().sort_values(ascending= False).head(1)
```

```python
# Find the best actor - genre combo in terms of avg metascore.
movies.groupby(["Star1", "Genre"])["Metascore"].mean().reset_index().sort_values(by= "Metascore", ascending= False)
```

```python
ipl = pd.read_csv("/content/deliveries.csv")
```

```python
ipl.head()
```

```python
ipl.shape
```

```python
# Find the top 10 batsman in terms of runs.
ipl.groupby("batsman")["batsman_runs"].sum().sort_values(ascending= False).head(10)
```

```python
# Find the batsman with maximum numbers of sixes.
sixes = ipl[ipl["batsman_runs"] == 6]
sixes.groupby("batsman")["batsman_runs"].value_counts().reset_index().sort_values(by= "count", ascending= False).head()
```

```python
# Find batsman with most number of 4's and 6's in the 5 over.
df = ipl[(ipl["over"] > 15) & ((ipl["batsman_runs"] == 4) | (ipl["batsman_runs"] == 6))]
df.groupby("batsman")["batsman"].count().sort_values(ascending= False)
```

```python
# Find Virat Kohli's record against all teams.
df = ipl[ipl["batsman"] == "V Kohli"]
df.groupby("bowling_team")["batsman_runs"].sum()
```

```python
# Create a function that returns the highest score of any batsman.
def highest_score(batsman):
  temp_df = ipl[ipl["batsman"] == str(batsman)]
  return temp_df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending= False).head(1)
```

```python
highest_score(batsman= "MS Dhoni")
```
